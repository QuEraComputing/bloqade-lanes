```python
from bloqade.lanes._prelude import kernel
from bloqade.lanes.arch.gemini.logical import get_arch_spec
from bloqade.lanes.bytecode.encoding import (
    LocationAddress,
    SiteLaneAddress,
    WordLaneAddress,
    ZoneAddress,
)
from bloqade.lanes.dialects import move
from bloqade.lanes.noise_model import generate_simple_noise_model
from bloqade.lanes.transform import MoveToSquinPhysical


word_lane = WordLaneAddress(0, 1, 0)
site_lane = SiteLaneAddress(1, 0, 0)


@kernel
def hand_routed_cz():
    state = move.load()

    state = move.fill(
        state,
        location_addresses=(
            LocationAddress(0, 0),
            LocationAddress(0, 1),
        ),
    )

    state = move.move(state, lanes=(word_lane,))
    state = move.move(state, lanes=(site_lane.reverse(),))

    state = move.cz(state, zone_address=ZoneAddress(0))

    state = move.move(state, lanes=(site_lane,))
    state = move.move(state, lanes=(word_lane.reverse(),))

    move.store(state)


arch_spec = get_arch_spec()

squin_kernel = MoveToSquinPhysical(
    arch_spec=arch_spec,
    noise_model=generate_simple_noise_model(),
).emit(hand_routed_cz)

squin_kernel.print()
```
